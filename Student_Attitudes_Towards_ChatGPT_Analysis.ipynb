### Function

```python
def print_uniques(df, cols=df.columns):
    for col in cols:
        print(f'{col}: {df[col].unique()}')
```

```python
def plot_pie_chart(data, labels, title):
    # Calculate the count of each unique value
    value_counts = data.value_counts()

    # Calculate total count
    total_count = value_counts.sum()

    # Create the pie chart
    plt.figure(figsize=(3, 3))

    # Customize labels for the pie chart
    legend_labels = [labels[label] if label in labels else str(label) for label in value_counts.index]

    # Add shadow for better visual appeal
    wedges, texts, autotexts = plt.pie(value_counts, colors=['skyblue', 'lightcoral'],
                                       autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, total_count * p / 100),
                                       startangle=90, shadow=True)

    plt.legend(wedges, legend_labels, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

    # Set custom colors for the text labels
    for autotext in autotexts:
        autotext.set_color('black')

    plt.title(title)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.show()
```

```python
def plot_horizontal_bar(data, title='', xlabel='', ylabel=''):
    # Calculate the total count for each label
    totals = data.sum()

    # Sort the totals DataFrame in descending order by values
    totals_sorted = totals.sort_values(ascending=True)

    # Create the chart
    plt.figure(figsize=(10, 4))
    bars = plt.barh(totals_sorted.index, totals_sorted.values, color='skyblue')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)

    # Display values on the right side of the bars
    for bar, value in zip(bars, totals_sorted):
        plt.text(value + 0.2, bar.get_y() + bar.get_height() / 2,
                 f'{value} ({value / data.shape[0] * 100:.1f}%)', va='center')

    plt.tight_layout()
    plt.show()
```

```python
def YearOfBirthError(data, column_name):
    list_error = []
    for i in data[column_name]:
        if len(i) != 4 or int(i) > 2020 or int(i) < 1950:
            list_error.append(i)
    return list_error
```

```python
def show_distribution(df, col_name, label='', title='', inline=True):
    col_counts = df[col_name].value_counts()

    # Create a 2-row 1-column grid
    if inline:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
    else:
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 9))

    # Set colors for the pie chart and autotext
    colors = ['skyblue', 'lightcoral', 'gold', 'lightgreen', 'orange']

    # Function to format the percentage values on the pie chart
    def format_autopct(pct):
        return f'{pct:.1f}%'

    wedges, texts, autotexts = ax1.pie(col_counts.values,
                                        autopct=format_autopct, colors=colors, startangle=0)

    # Set color and increase font size for autotext on the pie chart
    for autotext in autotexts:
        autotext.set_color('black')
        autotext.set_fontsize(12)

    # Add legend to the pie chart and place it outside the chart
    if label == '':
        label = col_name

    legend_labels = [f'{index}: {count}' for index, count in zip(col_counts.index, col_counts.values)]
    ax1.legend(wedges, legend_labels, title=label, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

    # Draw a countplot on the second subplot and round the columns
    sns.countplot(x=col_name, data=df, ax=ax2, palette=colors, order=col_counts.index)

    # Display integer counts on top of the countplot bars
    total_count = col_counts.sum()
    for p in ax2.patches:
        ax2.annotate(str(int(p.get_height())),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', fontsize=10, color='black',
                     xytext=(0, 10), textcoords='offset points')

    # Increase the spacing between the countplot bars and the frame
    ax2.margins(y=0.1)

    # Hide x and y labels on the countplot
    ax2.set_xlabel('')
    ax2.set_ylabel('')

    # Rotate the column labels by 45 degrees if the labels are too long
    if len(ax2.get_xticklabels()) > 3:
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')

    if(title == ''):
        plt.suptitle(f'Biểu đồ thống kê {col_name}')
    else:
        plt.suptitle(title)

    # Display the plot
    plt.tight_layout()
    plt.show()
```

##### Show Relationship

```python
def show_relationship(df, col_1, col_2, title=''):
    # Create a 1-row 2-column grid
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

    color_palette = ['skyblue', 'lightcoral', 'gold', 'lightgreen', 'orange']

    ### Plot on the first axis (Stacked Bar Plot)
    attitude = pd.crosstab(df[col_1], df[col_2])
    attitude_pct = attitude.div(attitude.sum(axis=1), axis=0)

    attitude_pct.plot(kind='bar', stacked=True, ax=ax1, color=color_palette)

    ax1.set_title('')
    ax1.set_xlabel('')
    ax1.set_ylabel('')

    if len(ax1.get_xticklabels()) > 4:
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
    else:
        ax1.set_xticklabels(ax1.get_xticklabels(), rotation=0, ha='center')

    for p in ax1.containers:
        for j, q in enumerate(p):
            height = q.get_height()
            width = q.get_width()
            x = q.get_x()
            y = q.get_y()
            x_center = x + width / 2
            y_center = y + height / 2
            x_cum = attitude_pct.iloc[:, :j].sum(axis=1).values
            x_text = x_cum[-1] + (x_center - x_cum[-1]) / 2
            ax1.annotate(f'{height:.1%}', (x_center, y_center),
                         xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

    ### Plot on the second axis (Count Plot)
    sns.countplot(x=col_1, hue=col_2, data=df, ax=ax2, palette=color_palette)
    ax2.set_title('')
    ax2.set_xlabel('')
    ax2.set_ylabel('')
    ax2.legend(title=col_2)

    # Add labels with counts above the bars
    for p in ax2.patches:
        height = p.get_height()
        ax2.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                     ha='center', va='bottom')

    if len(ax2.get_xticklabels()) > 4:
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')

    if title == '':
        plt.suptitle(f'Thống kê {col_2} theo {col_1}')
    else:
        plt.suptitle(title)

    plt.tight_layout()
    plt.show()
```

##### Show line plot

```python
def show_line_plot(data, col_1, col_2):
    # Calculate the proportion of each attitude category for each frequency category
    proportions = data.groupby([col_1, col_2]).size().unstack().div(data.groupby(col_1).size(), axis=0)

    # Sort the index and column labels of the proportions DataFrame in ascending order
    proportions_sorted = proportions.reindex(sorted(proportions.index), axis=0).reindex(sorted(proportions.columns), axis=1)

    # Create a line plot of the sorted proportions
    sns.lineplot(data=proportions_sorted, markers=True)
    plt.title(f'Mối quan hệ tỷ lệ giữa {col_1} và {col_2}')
    plt.xlabel(col_1)
    plt.ylabel('Tỷ lệ')
    plt.show()
```

```python
data = df.copy()
data[frequency_col].replace({'Dưới 1 lần/tuần': '0-Dưới 1 lần/tuần'}, inplace=True)
show_line_plot(data, frequency_col, attitude_col)
```
